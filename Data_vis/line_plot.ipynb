```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
phool=sns.load_dataset('iris')
phool
```

```python
sns.lineplot(x='sepal_length',y='sepal_width',data=phool)
plt.title('Phool')
plt.show()
```

# # Adding limits

```python
sns.lineplot(x='sepal_length',y='sepal_width',data=phool)
plt.title('Phool ')
plt.xlim(2)
plt.ylim(1)
plt.show()
```

# #Add style
- darkgrid
- whitegrid
- dark
- white
- ticks

```python
sns.set_style(style=None,rc=None)
sns.lineplot(x='sepal_length',y='sepal_width',data=phool)
plt.title('Phool')
sns.set_style('darkgrid')
plt.show()
```

```python
sns.lineplot(x='sepal_length',y='sepal_width',data=phool)
plt.title('Phool')
sns.set_style('dark')
sns.set_style(style=None,rc=None)
plt.show()
```

```python
sns.lineplot(x='sepal_length',y='sepal_width',data=phool)
plt.title('Phool')
sns.set_style('whitegrid')
plt.show()
```

```python
sns.set_style(style=None,rc=None)
sns.lineplot(x='sepal_length',y='sepal_width',data=phool)
plt.title('Phool')
sns.set_style('white')
plt.show()
```

```python
sns.set_style(style=None,rc=None)
sns.lineplot(x='sepal_length',y='sepal_width',data=phool)
plt.title('Phool')
sns.set_style('ticks')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(x='sepal_length',y='sepal_width',data=phool)
plt.title('Phool')
plt.show()
```
