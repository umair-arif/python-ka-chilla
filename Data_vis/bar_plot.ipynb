```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
phool=sns.load_dataset('iris')
phool
```

```python
sns.barplot(x='species',y='sepal_width',data=phool)
plt.title('Phool')
plt.show()
```

```python
sns.barplot(x='species',y='petal_length',data=phool)
plt.title('Phool')
plt.show()
```

```python
kashti=sns.load_dataset('titanic')
kashti
```

```python
sns.barplot(x='sex',y='alone',hue='who',data=kashti)
plt.title('Titanic')
plt.show()
```

```python
sns.barplot(x='sex',y='alone',hue='who',data=kashti,order=["female","male"])
plt.title('Titanic')
plt.show()
```

```python
sns.barplot(x='sex',y='alone',hue='who',data=kashti,color="red")
plt.title('Titanic')
plt.show()
```

```python
sns.barplot(x='sex',y='alone',hue='who',data=kashti,color="red",ci=None)
plt.title('Titanic')
plt.show()
```

```python
sns.barplot(x='sex',y='alone',hue='who',data=kashti,color="red",palette='pastel')
plt.title('Titanic')
plt.show()
```

```python
from numpy import median
import seaborn as sns
import matplotlib.pyplot as plt
kashti=sns.load_dataset('titanic')
sns.barplot(x="class",y="fare",hue="sex",data=kashti,estimator=median)
plt.title('Titanic')
plt.show()
```

```python
from numpy import median
import seaborn as sns
import matplotlib.pyplot as plt
kashti=sns.load_dataset('titanic')
sns.barplot(x="class",y="fare",hue="sex",data=kashti,estimator=median,saturation=2)
plt.title('Titanic')
plt.show()
```

```python
#horizontal barplot
from numpy import median
import seaborn as sns
import matplotlib.pyplot as plt
kashti=sns.load_dataset('titanic')
sns.barplot(x="fare",y="class",hue="sex",data=kashti,estimator=median,saturation=2)
plt.title('Titanic')
plt.show()
```

```python
from numpy import median
import seaborn as sns
import matplotlib.pyplot as plt
kashti=sns.load_dataset('titanic')
sns.barplot(x="class",y="fare",hue="sex",data=kashti,linewidth=3,facecolor=(0.1,0.3,0.3,0.5),errcolor='0.5',edgecolor='0.2')
plt.title('Titanic')
plt.show()
```
