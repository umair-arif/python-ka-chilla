```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks",color_codes=True)
```

```python
kashti=sns.load_dataset('titanic')
```

```python
kashti
```

```python
#countplot with 1 variable
p1=sns.countplot(x="sex",data=kashti)
```

```python
#countplot with 2 variable
p1=sns.countplot(x="sex",data=kashti,hue="class")
```

```python
#countplot with 2 variable and with title
p1=sns.countplot(x="sex",data=kashti)
p1.set_title("Plot for counting")
plt.show()
```

```python
#barplots
sns.catplot(x="sex",y="survived",hue="class",kind="bar",data=kashti)
plt.show()
```

```python
#continuous variables 
#Scatterplot:
g=sns.FacetGrid(kashti,row='sex',hue='alone')
g=g.map(plt.scatter,"age","fare")
```
