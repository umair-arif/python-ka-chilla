```python
series2 = pd.Series([3,5,7,9])
print(series2)
```

# Empty Series

```python
empty_s = pd.Series([])
print(empty_s)
```

# changing default index

```python
series3 = pd.Series([3,5,7,9], index=['a','b','c','d'])
print(series3)
```

# changing data type

```python
series3 = pd.Series([3,5,7,9], index=['a','b','c','d'], dtype=float)
print(series3)
```

# name parameter

```python
series3 = pd.Series([3,5,7,9], index=['a','b','c','d'], dtype=float , name='data values')
print(series3)
```

```python
scaler_s = pd.Series(0.5)
scaler_s
```

```python
scaler_s = pd.Series(0.5, index=[1,2,3])
scaler_s
```

```python
dict_s = pd.Series({'a':1,'b':2,'c':3})
dict_s
```

# Adding two series object

```python
s1 = pd.Series([1,2,3,4,5,6,7,8,9])
s2 = pd.Series([10,20,30,40,50,60,70,80,90])
```
